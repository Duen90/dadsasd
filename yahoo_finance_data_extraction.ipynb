```python
import yfinance as yf
import matplotlib.pyplot as plt
```

#### Extracting EUR-USD forex rates from yahoo finance with yfinance library

```python
data = yf.download("EURUSD=X", period="5d", interval="1m")

```

```python

plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.show()
```
